```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
data = pd.read_csv('Advertising.csv')
X = data.iloc[:, 1]
Y = data.iloc[:, 4]
plt.xlabel('TV')
plt.ylabel('Sales')
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
m = 0
c = 0


L = 0.00001 # The learning Rate
epochs = 50 # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X + c  #
    D_m = (-2/n) * sum(X * (Y - Y_pred))  
    D_c = (-2/n) * sum(Y - Y_pred)  
    m = m - L * D_m  
    c = c - L * D_c 

print (m, c)
```

```python
# Making predictions
Y_pred = m*X + c

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()


```

```python
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# read data into a DataFrame
dataxx = pd.read_csv('Advertising.csv', index_col=0)
dataxx.pop('Radio')
dataxx.pop('Newspaper')
dataxx.head()
```

```python
### STATSMODELS ###

# create a fitted model
lm1 = smf.ols(formula='Sales ~ TV', data=dataxx).fit()

# print the coefficients
lm1.params
```

```python
### SCIKIT-LEARN ###

# create X and y
feature_cols = ['TV']
X = dataxx[feature_cols]
y = dataxx.Sales

# instantiate and fit
lm2 = LinearRegression()
lm2.fit(X, y)

# print the coefficients
print(lm2.intercept_)
print(lm2.coef_)
```

```python
sns.pairplot(dataxx, x_vars=['TV'], y_vars='Sales', height=7, aspect=0.7, kind='reg')
```

```python
# Preprocessing Input data
XX = dataxx.iloc[:, 0]
YY = dataxx.iloc[:, 1]
plt.xlabel('TV')
plt.ylabel('Sales')
plt.scatter(XX, YY)
plt.show()
```

```python
Y_predx = lm2.coef_*XX +  lm2.intercept_
#Y_predx = 0.04753664*XX + 7.032593549127695

plt.scatter(XX, YY) 
plt.plot([min(XX), max(XX)], [min(Y_predx), max(Y_predx)], color='red')  # regression line
plt.show()
```
